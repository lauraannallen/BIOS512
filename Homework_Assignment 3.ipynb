Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.res = 250, repr.plot.height = 5, repr.plot.width = 5.5)

p = ggplot(penguins, aes(x = body_mass_g))+
    geom_histogram(color = 'deepskyblue4', fill = 'deepskyblue4', alpha = 0.75) +
    facet_wrap(~year, ncol = 1, scales = 'free_y') +
    theme_minimal() +
    labs(x = 'Body mass in grams', 
         y = 'Number of penguins',
        title = 'Histograms of penguin body mass for measurements taken from 2007-2009') +
    theme(
        plot.title = element_text(size = 9.5),
        strip.text = element_text(hjust = 0.9, face = 'bold', color = 'gray'))

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.res = 250, repr.plot.height = 6, repr.plot.width = 4.5)

p = ggplot(penguins, aes(x = body_mass_g, y = sex)) +
    geom_point(position = position_jitter(h = 0.2, w = 0), size = 1.5, alpha = 0.5, color = 'deepskyblue4') +
    facet_wrap(~species, ncol = 1, scales = 'free_y') +
    theme_minimal() + 
    theme(
        axis.title.y = element_blank(),
        strip.text = element_text(hjust = 0.9, face = 'bold', color = 'gray')) +
    labs(x = 'Body mass in grams')

p
```
