📍Hint: add `scale_color_brewer(type = "qual", palette = "Set2")` to adjust the colors of the second plot!

```r
options(repr.plot.width = 10)
p1 = ggplot(penguins, aes(x = body_mass_g, y = flipper_length_mm, color = species)) + geom_point(na.rm=TRUE)

p2 = ggplot(penguins, aes(x = body_mass_g, y = flipper_length_mm, color = island)) + 
    geom_point(na.rm=TRUE) +
    scale_color_brewer(type = "qual", palette = "Set2")
    
p1 + p2

# to change color plalete go to color brewer website and choose one 
```
